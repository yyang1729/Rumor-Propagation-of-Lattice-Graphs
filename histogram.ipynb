```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import statsmodels.api as sm
```

```python
y = []

n = "10000"
d = "5"

with open('Apexdata/path_N_' + n + "_D_" + d + '.csv') as fp:
    for line in fp:
        t = float(line)
        y.append(t)
with open('ndata/path_N_' + n + '_D_' + d + '.csv') as fp2:
  for line in fp2:
      t = float(line)
      y.append(t)

std = np.std(y)
mean = np.mean(y)

y = (y - mean) / std

def is_outlier(points):
  if len(points.shape) == 1:
        points = points[:,None]
  median = np.median(points, axis=0)
  diff = np.sum((points - median)**2, axis=-1)
  diff = np.sqrt(diff)
  med_abs_deviation = np.median(diff)

  modified_z_score = 0.6745 * diff / med_abs_deviation
  return modified_z_score > 3

y = y[~is_outlier(y)]

plt.title("N =" + n + ', d = ' + d + ', 1000 samples')
plt.hist(y, bins = 21)

ax2 = plt.twinx()
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax2.plot(x, stats.norm.pdf(x, mu, sigma), color = 'green')


plt.show()

fig = sm.qqplot(y, line='45')
```
